# Plots (Scatterplots)
* Latitude Vs Max Temperature
* Latitude Vs Humidity
* Latitude Vs Cloudiness
* Latitude Vs Wind Speed

# Using a Function for Plots

```python
#using a function to plot scatter plots

def plotting_data(x_axis, y_axis, x_lab, y_lab, title, savename):
    
    #scatterplot
    plt.scatter(x_axis, y_axis, marker = "o")
    plt.grid(True)

    #labeling
    plt.title(title)
    plt.xlabel(x_lab)
    plt.ylabel(y_lab)
    
    #sizing
    plt.tight_layout
    
    
    #exporting plot as png
    plt.savefig(f"../WebVisualizations/visualizations/{savename}.png")
    
    #printing
    plt.show()  
    
```

# Latitude Vs Max Temperature

```python
#labeling
x_lab = 'Latitude'
y_lab = 'Max Temperature (F)'
title = 'City Latitude vs. Max Temperature'

#saving
savename = 'fig 1 - Latitude_V_Max_Temperature'

#function
plotting_data(weather_df['Lat'], weather_df['Max Temp'], x_lab, y_lab, title, savename)
```

# Latitude vs. Humidity Plot

```python
#labeling
x_lab = 'Latitude'
y_lab = 'Humidity (%)'
title = 'City Latitude vs. Humidity'

#saving
savename = 'fig 2 - Latitude_V_Humidity'

#function
plotting_data(weather_df['Lat'], weather_df['Humidity'], x_lab, y_lab, title, savename)
```

# Latitude vs. Cloudiness Plot

# Latitude vs. Wind Speed Plot

```python
#labeling
x_lab = 'Latitude'
y_lab = 'wind Speed (mph)'
title = 'City Latitude vs. Wind Speed'

#saving
savename = 'fig 4 - Latitude_V_Wind_speed'

#function
plotting_data(weather_df['Lat'], weather_df['Wind Speed'], x_lab, y_lab, title, savename)
```
